### Fetching the url

```python
# the one we want ends with id=xxxx
# target the correct url using class .listBorder
url_class = doc.select('.listBorder')
url_class
```

```python
# creating the full list of first page with m.baobeihuijia.com
url_list = []
for link in url_class:
    list = link.select_one('a')['href']
    url_full = f"https://m.baobeihuijia.com/{list}"
    url_list.append(url_full)
```

### Pagination

```python
import time
```

```python
# first page is https://m.baobeihuijia.com/PhotoShow/list.aspx?tid=1&sex=&photo=&page=1
# last page is https://m.baobeihuijia.com/PhotoShow/list.aspx?tid=1&sex=&photo=&page=1477
# build a script that collects the link of each page
# i in range starts from 0, but since we plus one, our page ends at 1477.
# you can chain class .listBorder and <a> tag like this
url_full_list = []
time_start = time.time()

for page_num in range(0,10):
    url= f"https://m.baobeihuijia.com/PhotoShow/list.aspx?tid={page_num}"
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text, 'html.parser')
    
    for link in doc.select('.listBorder a'):
        link_url = link.get('href')
        url_complete = f"https://m.baobeihuijia.com/{link_url}"
        url_full_list.append(url_complete)
    print("up to page", page_num, "we found", len(url_full_list), "urls in total")
        
time_end = time.time()
print('time cost', time_end - time_start, 's')
```
